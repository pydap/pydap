## ¿Esta es una URL DAP2 o DAP4?




Solo dos servidores `OPeNDAP` soportan ampliamente `DAP4`: [Hyrax Data Server](https://www.opendap.org/software/hyrax-data-server/) (desarrollado por [OPeNDAP Inc](https://www.opendap.org/), los arquitectos originales de `OPeNDAP`) y [Thredds Data Server](https://www.unidata.ucar.edu/software/tds/) (desarrollado por [NSF Unidata](https://www.unidata.ucar.edu/)). `DAP4` fue desarrollado originalmente de forma conjunta por `OPeNDAP` y `Unidata` en 2016 (consulta [la especificación DAP4](https://opendap.github.io/dap4-specification/DAP4.html)). Cualquier servidor `Hyrax` y `Thredds` que implemente `DAP4` también implementa `DAP2`, y comprobar si una URL es DAP4 o DAP2 es diferente en cada uno de estos servidores.

### <font size="5"><span style='color:#0066cc'> **Hyrax Data Server**<font size="3">
Con `Hyrax`, la forma más sencilla de verificar si la URL soporta DAP4 o no es tomar la URL, agregar `.dmr` (o `.dmr.html`) y pegarla en un navegador. Por ejemplo, considera la siguiente URL opendap de Hyrax:

```python
URL = http://test.opendap.org/opendap/data/nc/coads_climatology.nc
```

| **URL DAP2 Hyrax** | **URL DAP4 Hyrax** |
| :- |  -: |
| http://test.opendap.org/opendap/data/nc/coads_climatology.nc.dds | http://test.opendap.org/opendap/data/nc/coads_climatology.nc.dmr
| http://test.opendap.org/opendap/data/nc/coads_climatology.nc.html | http://test.opendap.org/opendap/data/nc/coads_climatology.nc.dmr.html
| http://test.opendap.org/opendap/data/nc/coads_climatology.nc.das | http://test.opendap.org/opendap/data/nc/coads_climatology.nc.dmr.xml


```{note}
Un servidor Hyrax que implementa DAP4 debería implementar DAP2. En la mayoría de los servidores Hyrax (versión > 1.15), DAP4 es la implementación de protocolo predeterminada, y cuando una persona usuaria solicita una respuesta DAP2, el servidor primero procesa y luego traduce una respuesta DAP4 a DAP2. Esto habilita compatibilidad hacia atrás con clientes antiguos que solo pueden analizar respuestas DAP2, a la vez que soporta datasets más nuevos. En ese proceso, es posible que el dataset remoto no pueda traducirse por completo a una respuesta DAP2, y Hyrax puede devolver un Error 400 en su lugar (<a href="Q4.html#a-word-of-caution">consulta la advertencia</a>).
```

```{warning}
Para ver la versión oficial más reciente de `Hyrax`, ve al [sitio oficial de OPeNDAP](https://www.opendap.org/) y revisa `Latest Release` bajo `Software`.
```



### <font size="5"><span style='color:#0066cc'> **Thredds Data Server**<font size="3">

Con `TDS`, una URL `DAP2` tendrá `dodsC` en su URL. Si reemplazas `dodsC` por `dap4` y la pegas en un navegador, debería producir una página web que permite descargar una respuesta DAP4. Por ejemplo, considera:

```python
URL = https://thredds-test.unidata.ucar.edu/thredds/dodsC/testdods/coads_climatology.nc
```


| **URL DAP2 Thredds** | **URL DAP4 Thredds** |
| :- |  -: |
| https://thredds-test.unidata.ucar.edu/thredds/dodsC/testdods/coads_climatology.nc.html | https://thredds-test.unidata.ucar.edu/thredds/dap4/testdods/coads_climatology.nc.html
| https://thredds-test.unidata.ucar.edu/thredds/dodsC/testdods/coads_climatology.nc.dds | https://thredds-test.unidata.ucar.edu/thredds/dap4/testdods/coads_climatology.nc.dmr.xml |

```{warning}
Un servidor `Thredds` que implementa DAP2 puede no implementar DAP4. Se recomienda firmemente comprobar si un servidor está configurado para ofrecer el endpoint DAP4.
```


```{note}
Para ver la versión oficial más reciente de `Thredds`, ve a la [página oficial de Thredds](https://www.unidata.ucar.edu/software/tds/).
```
