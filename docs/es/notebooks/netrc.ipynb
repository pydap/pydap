## Crear un archivo .netrc

La ubicación predeterminada de este archivo es el directorio home, de modo que solo tú puedas usarlo. Esto es:
- LinuxOS / MacOS : `~/.netrc`
- Windows OS: `C:\Users\<User-Name>/_netrc`.
```{note}
En el resto de este notebook asumiremos que tienes LinuxOS / MacOS.
```
### Pasos para configurar tu archivo .netrc
1. Créalo en la ubicación predeterminada
```
touch ~/.netrc
```
2. Agrega credenciales de autenticación (por ejemplo, Earth Data Login), usando cualquiera de los siguientes formatos:

```
machine urs.earthdata.nasa.gov
login <YourEDLUserName>
password <YourEDLPassword>
```

o

```
machine urs.earthdata.nasa.gov login <YourEDLUserName> password <YourEDLPassword>
```

```{note}
Tu `.netrc` puede contener credenciales de autenticación fuera del ámbito de NASA. Por ejemplo, si tienes una cuenta de AVISO Altimetry y quieres acceder a datos TDS, puedes agregar otra entrada a tu archivo `.netrc`, como: `machine tds.aviso.altimetry.fr login <YourAvisoUsername> password <YourAvisoPassword>`
```

3. Otorga permisos a tu archivo .netrc:
```
chmod 600 ~/.netrc
```
4. [Opcional] Prueba que se pueda usar con una herramienta de línea de comandos. Por ejemplo, con curl:

```
curl -L -n <OPeNDAP EarthData URL> 
```
donde `-L` implica seguir redirecciones, y `-n` le indica a curl que busque el archivo netrc en la ubicación predeterminada.
